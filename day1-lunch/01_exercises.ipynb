# Exercise: Modify the introductory plot so that each subplot shares the x- and y-axis scales.

```python
import matplotlib.pyplot as plt
```

## Original plots

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
x2 = [2, 4, 6]
y2 = [8, 64, 216]
```

```python
fig, ax = plt.subplots( nrows=2 )
ax[0].plot( x, y, '.' )
ax[1].plot( x2, y2 )
plt.show()
```

## Modified plots

```python
# Create plots
fig, ax = plt.subplots( nrows=2 )
ax[0].plot( x, y, '.' )
ax[1].plot( x2, y2 )

# Modify x-axis scale on both plots
ax[0].set_xlim(0, 6)
ax[1].set_xlim(0, 6)

# Modify y-axis scale on both plots
ax[0].set_ylim(0, 220)
ax[1].set_ylim(0, 220)

plt.show()
```
