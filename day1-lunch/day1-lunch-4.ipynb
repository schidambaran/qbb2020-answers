## 4. Plot K4, K9, and K27 width distributions on the same plot as subplots.

```python
# Load libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Load data

```python
# Name columns
bed_columns = ["chr", "start", "end"]

# Load in K4 data and add column for widths
k4_data = pd.read_table("/Users/cmdb/qbb2020/data/K4me3.bed", names = bed_columns)
k4_data.loc[:, "width"] = k4_data.loc[:, "end"] - k4_data.loc[:, "start"]

# Load in K9 data and add column for widths
k9_data = pd.read_table("/Users/cmdb/qbb2020/data/K9me3.bed", names = bed_columns)
k9_data.loc[:, "width"] = k9_data.loc[:, "end"] - k9_data.loc[:, "start"]

# Load in K27 data and add column for widths
k27_data = pd.read_table("/Users/cmdb/qbb2020/data/K27me3.bed", names = bed_columns)
k27_data.loc[:, "width"] = k27_data.loc[:, "end"] - k27_data.loc[:, "start"]
```

### Plot width distributions

```python
fig, axes = plt.subplots(ncols = 3, figsize = (10, 4))

# Plot K4 width distribution
axes[0].hist(k4_data.loc[:, "width"])
axes[0].set_title("K4 Width Distribution")
axes[0].set_xlabel("Width")
axes[0].set_ylabel("Frequency")
axes[0].tick_params(rotation = 45)

# Plot K9 width distribution
axes[1].hist(k9_data.loc[:, "width"])
axes[1].set_title("K9 Width Distribution")
axes[1].set_xlabel("Width")
axes[1].set_ylabel("Frequency")
axes[1].tick_params(rotation = 45)

# Plot K27 width distribution
axes[2].hist(k27_data.loc[:, "width"])
axes[2].set_title("K27 Width Distribution")
axes[2].set_xlabel("Width")
axes[2].set_ylabel("Frequency")
axes[2].tick_params(rotation = 45)

plt.tight_layout()
plt.show()
```
