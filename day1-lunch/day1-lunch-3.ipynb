## 3. Plot K4, K9, and K27 chromosome distributions on the same plot as subplots.

```python
# Load libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Load data

```python
# Name columns
bed_columns = ["chr", "start", "end"]

# Load in K4 data
k4_data = pd.read_table("/Users/cmdb/qbb2020/data/K4me3.bed", names = bed_columns)

# Load in K9 data
k9_data = pd.read_table("/Users/cmdb/qbb2020/data/K9me3.bed", names = bed_columns)

# Load in K27 data
k27_data = pd.read_table("/Users/cmdb/qbb2020/data/K27me3.bed", names = bed_columns)
```

### Sort chromosomes alphanumerically

```python
# Format and sort K4 chromosome distribution data
k4_chr = k4_data.loc[:, "chr"]
k4_chr = k4_chr.value_counts()
k4_chr = k4_chr.sort_index()

# Format and sort K9 chromosome distribution data
k9_chr = k9_data.loc[:, "chr"]
k9_chr = k9_chr.value_counts()
k9_chr = k9_chr.sort_index()

# Format and sort K27 chromosome distribution data
k27_chr = k27_data.loc[:, "chr"]
k27_chr = k27_chr.value_counts()
k27_chr = k27_chr.sort_index()
```

### Plot chromosome distributions

```python
fig, axes = plt.subplots(ncols = 3, figsize = (10, 4))

# Plot K4 chromosome distribution
axes[0].bar(k4_chr.index, k4_chr)
axes[0].set_title("K4 Chromosome Distribution")
axes[0].set_xlabel("Chromosome")
axes[0].set_ylabel("Frequency")
axes[0].tick_params(rotation = 45)

# Plot K9 chromosome distribution
axes[1].bar(k9_chr.index, k9_chr)
axes[1].set_title("K9 Chromosome Distribution")
axes[1].set_xlabel("Chromosome")
axes[1].set_ylabel("Frequency")
axes[1].tick_params(rotation = 45)

# Plot K27 chromosome distribution
axes[2].bar(k27_chr.index, k27_chr)
axes[2].set_title("K27 Chromosome Distribution")
axes[2].set_xlabel("Chromosome")
axes[2].set_ylabel("Frequency")
axes[2].tick_params(rotation = 45)

plt.tight_layout()
plt.show()
```
